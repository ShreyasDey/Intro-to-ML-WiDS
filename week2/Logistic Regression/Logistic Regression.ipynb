```python
import numpy as np
import pandas as pd
import copy
```

```python
df=pd.read_csv('framingham.csv')
```

```python
df = df.replace({'NA': 0})
df=df.fillna(0)
```

```python
print(df)
```

```python
df=df.to_numpy()
```

```python
print(df)
```

```python
train=df[:3814,:]
X_train=train[:,:-1]
y_train=train[:,-1]
test=df[3814:,:]
print(y_train)
```

```python
def sigmoid(x):
  return 1 / (1 + np.exp(-x))
```

```python
def gradient_descent(X, y, w_in, b_in, gradient_function, alpha, num_iters): 
    w = copy.deepcopy(w_in)
    b = b_in   
    for i in range(num_iters):
        dj_db,dj_dw = gradient_function(X, y, w, b)
        w = w - alpha * dj_dw
        b = b - alpha * dj_db     
    return w, b
```

```python
def compute_gradient(X, y, w, b): 
    m,n = X.shape
    dj_dw = np.zeros((n,))
    dj_db = 0.
    for i in range(m):
        f_wb_i = sigmoid(np.dot(X[i],w) + b)
        err_i  = f_wb_i  - y[i]
        for j in range(n):
            dj_dw[j] = dj_dw[j] + err_i * X[i,j]
        dj_db = dj_db + err_i
    dj_dw = dj_dw/m
    dj_db = dj_db/m        
    return dj_db, dj_dw
```

```python
w_final, b_final = gradient_descent(X_train, y_train, list(np.zeros(15)+10),10.0, compute_gradient, 5.0e-7, 1000)
print(f"b,w found by gradient descent: {b_final},{w_final} ")
```

```python
X_test=test[:,:-1]
y_test=test[:,-1]
print(y_test)
```

```python
result=[]
for i in X_test:
    if np.dot(w_final,i)+b_final>=0.5:
        result.append(1)
    else:
        result.append(0)
result=np.array(result)
```

```python
print(result)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, result)
print("Confusion Matrix:\n", cm)
```
