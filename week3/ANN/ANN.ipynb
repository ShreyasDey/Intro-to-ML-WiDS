```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('updated_air_quality_data.csv')
```

```python
df
```

```python
df=df[df['Geo Place Name']!='Unknown']
```

```python
print(str(df['Data Value'].dtype))
```

```python
columns_to_encode=[]
for i in list(df.columns.to_numpy()):
    if str(df[i].dtype)!='int64' and str(df[i].dtype)!='float64':
        columns_to_encode.append(i)
```

```python
columns_to_encode.remove('Start_Date')
```

```python
columns_to_encode
```

```python
# Dictionary to store the fitted encoders for later use
encoders = {}
for column in columns_to_encode:
    # Create a new encoder for each column
    le = LabelEncoder()
    # Fit and transform the data, overwriting the original column
    df[column] = le.fit_transform(df[column])
    # Save the fitted encoder
    encoders[column] = le
print("\nEncoded DataFrame:")
print(df)
```

```python
df['Start_Date'] = pd.to_datetime(df['Start_Date'])

# Create new columns for year, month, and day
df['Year'] = df['Start_Date'].dt.year
df['Month'] = df['Start_Date'].dt.month
df['Day'] = df['Start_Date'].dt.day

# Print the resulting DataFrame
df.drop('Start_Date',axis=1,inplace=True)
```

```python
df
```

```python
X=df.drop(['Data Value','Air Quality Category'],axis=1)
y=df[['Data Value','Air Quality Category']]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.15,random_state=42)
```

```python
y_dv_train=y_train['Data Value']
y_aqc_train=y_train['Air Quality Category']
y_dv_test=y_test['Data Value']
y_aqc_test=y_test['Air Quality Category']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test  = scaler.transform(X_test)
```

```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import BinaryCrossentropy,SparseCategoricalCrossentropy
```

```python
dv_model = Sequential([
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(32,activation='relu'),
    Dense(16,activation='relu'),
    Dense(8,activation='relu'),
    Dense(4,activation='relu'),
    Dense(2,activation='relu'),
    Dense(1, activation='linear')
])

dv_model.compile(optimizer=Adam(learning_rate=1e-3),loss=BinaryCrossentropy(),metrics=['mse'])
dv_model.summary()
```

```python
dv_model.fit(X_train,y_dv_train,validation_split=15/85,epochs=50,batch_size=32,verbose=1)
```

```python
aqc_model = Sequential([
    Dense(128, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(64, activation='relu'),
    Dense(3, activation='linear')
])

aqc_model.compile(optimizer=Adam(learning_rate=1e-3),loss=SparseCategoricalCrossentropy(from_logits=True),metrics=['accuracy'])

aqc_model.summary()
```

```python
aqc_model.fit(X_train,y_aqc_train,validation_split=0.1765,epochs=50,batch_size=32,verbose=1)
```

```python
dv_mse = dv_model.evaluate(X_test, y_dv_test, verbose=0)
print(f"Test MSE (Data Value): {dv_mse[1]}")
```

```python
aqc_loss, aqc_acc = aqc_model.evaluate(X_test, y_aqc_test, verbose=0)
print(f"Test Accuracy (Air Quality Category): {aqc_acc}")
```
