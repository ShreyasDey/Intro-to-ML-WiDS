```python
import tensorflow as tf
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from sklearn.preprocessing import OneHotEncoder
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Dropout,Conv2D,Flatten,Reshape
from tensorflow.keras.optimizers import Adam
```

```python
mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
# After doing the data processing, one hot encoding etc , build a CNN model and measure its accuracy and try to optimize the mode
```

```python
y_test
```

```python
encoder=OneHotEncoder()
y=np.array(encoder.fit_transform(y_train.reshape(-1,1)).toarray())
yt=np.array(encoder.fit_transform(y_test.reshape(-1,1)).toarray())
x_train[0].shape
```

```python
yt
```

```python
tf.random.set_seed(1234) # for consistent results
```

```python
model = Sequential(
    [   
        Reshape((28, 28, 1), input_shape=(28, 28)),
        Dense(16,activation='relu'),
        Conv2D(32,kernel_size=(3,3),activation='relu'),
        Conv2D(64,kernel_size=(3,3),activation='relu'),
        Conv2D(128,kernel_size=(3,3),activation='relu'),
        Flatten(),
        Dense(10,activation='softmax')
    ]
)
model.compile(loss='categorical_crossentropy',optimizer=Adam(learning_rate=1e-3),metrics=['accuracy'])
model.summary()
```

```python
model.fit(x_train,y,epochs=5,batch_size=32,verbose=1)
```

```python
modeld = Sequential(
    [   
        Reshape((28, 28, 1), input_shape=(28, 28)),
        Dense(16,activation='relu'),
        Conv2D(32,kernel_size=(3,3),activation='relu'),
        Dropout(0.5),
        Conv2D(64,kernel_size=(3,3),activation='relu'),
        Dropout(0.5),
        Conv2D(128,kernel_size=(3,3),activation='relu'),
        Flatten(),
        Dense(10,activation='softmax')
    ]
)
modeld.compile(loss='categorical_crossentropy',optimizer=Adam(learning_rate=1e-3),metrics=['accuracy'])
modeld.summary()
```

```python
modeld.fit(x_train,y,epochs=5,batch_size=32,verbose=1)
```

```python
model_loss, model_acc = model.evaluate(x_test, yt, verbose=1)
print(f"Accuracy without dropout: {model_acc}")
```

```python
modeld_loss, modeld_acc = modeld.evaluate(x_test, yt, verbose=1)
print(f"Accuracy with dropout: {modeld_acc}")
```

Accuracy is coming better with dropout than without dropout
